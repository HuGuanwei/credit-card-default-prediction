```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ada = pd.read_csv('../data/ada.csv').values[:,1]
rf = pd.read_csv('../data/rf.csv').values[:,1]
nn = pd.read_csv('../data/nn.csv').values
ensemble = pd.read_csv('../data/ensemble.csv').values
y = pd.read_csv('../data/y_test.csv').values
```

```python
from sklearn.metrics import roc_curve, auc
from sklearn import metrics

def plot_auc(y_prob, y, model_name, color):
    plt.title('Receiver Operating Characteristic')
    fpr, tpr, threshold = metrics.roc_curve(y, y_prob)
    roc_auc = metrics.auc(fpr, tpr)
    plt.plot(fpr, tpr, c = color, label = model_name + ' AUC = %0.9f' % roc_auc)
    plt.legend()
    plt.plot([0, 1], [0, 1],'r--')
    plt.xlim([0, 1])
    plt.ylim([0, 1])
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    return
```

```python
plt.figure(figsize=(8, 4.5))
plot_auc(ada, y, 'AdaBoost', 'royalblue')
plot_auc(rf, y, 'RandomForest', 'seagreen')
plot_auc(nn, y, 'NeuralNetwork', 'gold')
plot_auc(ensemble, y, 'Ensemble', 'orange')
```
